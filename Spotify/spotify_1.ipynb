# Preparing my pipeline

```python
num_col = ['acousticness','danceability',
           'duration_ms','energy','instrumentalness',
           'liveness','loudness','speechiness',
           'tempo','valence']

cat_col = ['mode','time_signature','key']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler, OneHotEncoder

from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression, Ridge, LinearRegression
from sklearn.model_selection import cross_val_score, GridSearchCV

num_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())])

cat_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse=False, drop='if_binary'))])

preprocessor = ColumnTransformer([
    ('num_tr', num_transformer, num_col),
    ('cat_tr', cat_transformer, cat_col)], remainder='drop')
```

```python
pipe = Pipeline([
    ('preproc', preprocessor),
    ('model', LinearRegression())
])
```

```python
SimpleImputer.get_feature_names_out = (lambda self, names=None: self.feature_names_in_)
```

```python
from sklearn.model_selection import train_test_split
```

```python
def vif_table(df, features, with_standardize=False):
    if with_standardize:
        df = standardize(df, features)
    vif_table = pd.DataFrame()
    vif_table['index'] = [key for index, key in enumerate(df.keys())]
    vif_table['vif'] = [
        vif(df.values, index) for index, key in enumerate(df.keys())
    ]
    return vif_table
```

```python
def scat_plot(x, y, hue=None, xlab='', ylab='', titl=''):
    '''Plots a scatterplot using given inputs'''
    fig, ax = plt.subplots(figsize=(10,6))
    sns.scatterplot(x, y, hue=hue, s=12)
    plt.xlabel(xlab, fontsize=12)
    plt.ylabel(ylab, fontsize=12)
    plt.title(titl, fontsize=14)
    plt.legend(fontsize=12)
    plt.show()
    
def regress_plot(x='', y='', data=None, xlab='', ylab='', titl=''):
    '''Plots a scatterplot with a regression line
    using given inputs'''
    fig, ax = plt.subplots(figsize=(10,6))
    sns.regplot(x, y, data=data, scatter_kws={"s": 10}, line_kws={'color':'r'})
    plt.xlabel(xlab, fontsize=12)
    plt.ylabel(ylab, fontsize=12)
    plt.title(titl, fontsize=14)
    plt.ylim(-3, 103)
    plt.show()
```
